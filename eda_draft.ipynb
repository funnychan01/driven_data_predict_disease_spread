```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
```

```python
input_path = 'D:\\Jackie\\Documents\\2. Data Science Projects\\20181025_PredictDiseaseSpread\\2. Prepared Data\\'
```

```python
raw_df = pd.read_csv(input_path + 'dengue_features_train.csv')
raw_df.fillna(0, inplace=True)
raw_label = pd.read_csv(input_path + 'dengue_labels_train.csv')
X_train, X_test, y_train, y_test = train_test_split(raw_df, raw_label, test_size=0.3, random_state=18)
```
